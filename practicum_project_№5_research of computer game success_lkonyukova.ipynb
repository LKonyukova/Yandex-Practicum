Нужно посмотреть на пропуски детальнее и решить, что с ними делать. Особенно в столбцах с оценкой и рейтингом, 40-50% пропусков  — это слишком много. Что интересно, бОльшая часть пропусков сосредоточена в первые годы.

Сначала посмотрим на пропуски в `name`.

```python
games.loc[games['name'].isna()]
```

В этих же строках и наши пропуски из столбца `genre`.
Похоже мы никак не идентифицируем эти игры, поэтому удалим строки.

```python
games = games.dropna(subset=['name'])

# проверим себя
games['name'].isna().sum()
```

Теперь посмотрим на пропуски в `year_of_release`.

```python
games.loc[games['year_of_release'].isna()].head(10)
```

Сложно предположить, можем ли мы как-то восстановить год. Только если искать дату релиза каждой конкретной игры на конкретной платформе. Кажется, игра не стоит свеч. Поэтому удалим эти строки, т.к. их не так много. 

```python
games = games.dropna(subset=['year_of_release'])

# проверим себя
games['year_of_release'].isna().sum()
```

Посмотрим на данные в столбце `critic_score`. Согласно документации там должны быть оценки по 100-балльной системе.

```python
display(games['critic_score'].unique())

print('Количество пропусков:', games['critic_score'].isna().sum())
```

Удалить больше 51% данных мы не можем. Заменить пустые клетки медианным или средним значением было бы тоже не правильно, т.к. это сильно исказит данные. Заменим Nan на заглушку -1.

```python
games['critic_score'] = games['critic_score'].fillna(value=-1)

# проверим себя
games['critic_score'].isna().sum()
```

Теперь посмотрим на данные в столбце `user_score`. Согласно документации там должны быть оценки по 10-балльной системе.

```python
display(games['user_score'].unique())

print('Количество пропусков:', games['user_score'].isna().sum())
```

У нас есть tbd значения (to be determined = будет определено). Значит оценка пользователями там ещё не выставлена, но для нас эти значения всё равно что отсутствующие. Т.к. в этом столбце тоже большой объём пропущенных значений, то просто удалить эти строки мы также не можем. Заменим tbd и Nan, как и ранее, на заглушку -1.

```python
games['user_score'] = games['user_score'].fillna(value=-1)
games['user_score'] = games['user_score'].replace('tbd', '-1')

# проверим себя
games['critic_score'].isna().sum()
```

И наконец поработаем с пропусками в столбце `rating`. Там хранится рейтинг игры в текстовом формате. Вероятнее всего, если рейтинг не указан, его ещё не успели выставить. Значит пустые значения заменим на "not_rated" (не оценено).

```python
games['rating'] = games['rating'].fillna(value='not_rated')

# проверим себя
games['rating'].isna().sum()
```

Подумаем о природе пропусков. Посмотрим на данные.

```python
display(games.query('rating == "not_rated"').sort_values(by='year_of_release').head(15))
games.query('rating == "not_rated"').sort_values(by='year_of_release').tail(15)
```

Большее количество пропусков одновременно встречается в трёх столбцах (`critic_score`, `user_score` и `rating`). Можно выдвинуть предположение, что данная проблема встречается только у старых игр и что ранее их так не оценивали.\
И правда, организация ESRB была основана в 1994 г, что объясняет практически полное отсутствие рейтинга до этого года вовсе.
Но важно отметить, что если добавить фильтрацию по году выпуска игры, мы видим, что данная проблема встречается как и у старых игр (1980-х), так и у более современных (вплоть до 2016 года включительно). Так что такое предположение мы не можем подтвердить в полной мере.

Из более вероятно почему нет рейтингов и оценок у более современных игр, можно предположить, что:
- в процессе скачивания данных произошла техническая ошибка;
- оценка и/или рейтинг не стоят у не популярных игр и/или на не популярной консоли, т.к. не дошли до необходимого порога по количеству оценок для отображения рейтинга;
- игра может быть востребована, например в Японии, где более популярные иные порталы для оценки игр, так что местные пользователи склонны выставалять рейтинг именно там.

```python
display(games.isna().sum())

print('Размер датасета после обработки:', games.shape)
```

Отлично, у нас не осталось пропусков.\
В ходе обработки мы потеряли всего 1,6% исходных данных.

Посмотрим на уникальные названия градаций рейтинга и сравним их с официальной расшифровкой рейтинга ESRB.

Расшифровка рейтинга:
* **EC** («Early childhood») — для детей младшего возраста
* **E** («Everyone») — для всех
* **E10+** («Everyone 10 and older») — для всех от 10 лет и старше
* **T** («Teen») — для подростков
* **M** («Mature») — для взрослых
* **AO** («Adults Only 18+») — только для взрослых
* **RP** («Rating Pending») — рейтинг ожидается

```python
games['rating'].unique()
```

У нас есть неопознанный рейтинг `K-A`.\
Из данных [сети](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board#:~:text=%D0%9F%D0%B5%D1%80%D0%B2%D0%BE%D0%BD%D0%B0%D1%87%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%20%22K%2DA%22%20(%22Kids,ESRB%202%20%D0%BC%D0%B0%D1%80%D1%82%D0%B0%202005%20%D0%B3%D0%BE%D0%B4%D0%B0.) мы узнаём, что `K-A` («Kids to Adults») — это первоначальная версия `E` («Everyone») — для всех.\
Произведём замену `K-A` на `E` для унификации.

```python
games['rating'] = games['rating'].replace('K-A', 'E')
```

```python
games['rating'] = games['rating'].replace('RP', 'not_rated')

# проверим себя
games['rating'].unique()
```

Нам повезло с данными, в ходе обработки удалили только одну дублирующую строку и обновили рейтинг, а в остальном всё хорошо. Можно идти дальше.

## Проверим гипотезы

Нам нужно проверить две гипотезы:
- Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
- Гипотеза 2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

И в первом, и во втором случае у нас две независимые выборки по которым мы должным проверить гипотезу о развенстве средних двух генеральных совокупностей. Поэтому будем в обоих случаех иcпользовать метод **scipy.stats.ttest_ind()**.

Уровень стат. значимости установим 0,5%, т.к. у нас нет необходимости в высокой точности.

### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

<b>Выдивинем гипотезы:</b>\
<b>H0</b> : μ1 = μ2, средние пользовательские рейтинги платформ Xbox One и PC одинаковые.\
<b>H1</b> : μ1 ≠ μ2, средние пользовательские рейтинги рейтингов платформ Xbox One и PC различаются.

```python
# в датасете games_actual_score_user данные за период с 2013 по 2016 и уже убраны заглушки -1 
games_actual_XOne = games_actual_score_user.query('platform == "XOne"')
games_actual_XOne['user_score']
games_actual_PC = games_actual_score_user.query('platform == "PC"')
games_actual_PC['user_score']

alpha = 0.05

# параметр alternative не указываем, т.к. нам нужно значеник "two-sided", идущее по умолчанию
results = st.ttest_ind(games_actual_XOne['user_score'], games_actual_PC['user_score'])

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу => средние пользовательские рейтинги рейтингов платформ Xbox One и PC различаются')
else:
    print('Нет оснований отвергнуть нулевую гипотезу => средние пользовательские рейтинги платформ Xbox One и PC одинаковые')
```

Проверим себя и посмотрим на средние пользовательские рейтинги по платформам.

```python
print('Средние пользовательские рейтенги Xbox One:', games_actual_XOne['user_score'].mean())
print('Средние пользовательские рейтенги PC:', games_actual_PC['user_score'].mean())
```

<b>Вывод:</b> есть основания говорить, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.\
Гипотеза 1 подтвердилась.

### Гипотеза 2. Средние пользовательские рейтинги жанров Action и Sports разные

<b>Выдивинем гипотезы:</b>\
<b>H0</b> : μ1 = μ2, средние пользовательские рейтинги жанров Action и Sports одинаковые.\
<b>H1</b> : μ1 ≠ μ2, средние пользовательские рейтинги жанров Action и Sports различаются.

```python
# в датасете games_actual_score_user данные за период с 2013 по 2016 и уже убраны заглушки -1 
games_actual_Action = games_actual_score_user.query('genre == "Action"')
games_actual_Action['user_score']
games_actual_Sports = games_actual_score_user.query('genre == "Sports"')
games_actual_Sports['user_score']

alpha = 0.05

# параметр alternative не указываем, т.к. нам нужно значение "two-sided", идущее по умолчанию
results = st.ttest_ind(games_actual_Action['user_score'], games_actual_Sports['user_score'])

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу => средние пользовательские рейтинги жанров Action и Sports различаются')
else:
    print('Нет оснований отвергнуть нулевую гипотезу => средние пользовательские рейтинги жанров Action и Sports одинаковые')
```

Проверим себя и посмотрим на средние пользовательские рейтинги по жанрам.

```python
print('Средние пользовательские рейтенги Action:', games_actual_Action['user_score'].mean())
print('Средние пользовательские рейтенги Sports:', games_actual_Sports['user_score'].mean())
```

<b>Вывод:</b> есть основания говорить, что средние пользовательские рейтинги жанров Action и Sports различаются.\
Гипотеза 2 подтвердилась.

В Японии трудно определить самый популярный рейтинг, т.к. у них на первом месте игры, у которых нет рейтинга (60,4%). Оставшиеся 40% были разделены между играми для подростков (`T`), играми для всех (`E`), играми для взрослых (`M`) и играми для всех от 10 и старше (`E10+`). Интересно отметить, что играм для взрослых (`M`), столь популярным в Северной Америке и Европе, в Японии отведено лишь 10%.
