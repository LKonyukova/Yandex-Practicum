Посмотриим, есть ли зависимости пропусков от данных. Для этого по данным столбцов `age` и `balance` создадим новые столбцы `age_missing` и `balance_missing` с пропущенными значениями.

```python
# создадим столбцы
data['age_missing'] = data['age'].isnull()#.astype(int)
data['balance_missing'] = data['balance'].isnull()#.astype(int)

age_pivot = data.groupby('age_missing').agg({'equity':'median',
                                             'churn':'median',
                                             'products':'median',
                                             'credit_card':'median',
                                             'last_activity':'median',
                                             'age':'mean', 
                                             'score':'mean', 
                                             'balance':'mean', 
                                             'est_salary':'mean'
                                             }).round(2).transpose()

age_pivot

balance_pivot = data.groupby('balance_missing').agg({'equity':'median',
                                             'churn':'median',
                                             'products':'median',
                                             'credit_card':'median',
                                             'last_activity':'median',
                                             'age':'mean', 
                                             'score':'mean', 
                                             'balance':'mean', 
                                             'est_salary':'mean'
                                             }).round(2).transpose()

# выведем результат
print('Анализ пропусков в столбце "age", где False - нет пропусков, а True - есть')
display(age_pivot)
print('-'*78)
print('Анализ пропусков в столбце "balance", где False - нет пропусков, а True - есть')   
display(balance_pivot)
```

**Вывод:** мы видим, что для строк с пропущенными значениями в столбце `age` значительно отличаются показатели `balance` и `equity` от строк, где возраст указан.

## Шаг 4: Проверка статистических гипотез

### Гипотеза 1

Сформулируем первую пару гипотез: 

- **H0** : μ1 = μ2, Между доходом ушедших и оставшихся клиентов **нет** статистически значимой разницы.
- **H1** : μ1 ≠ μ2, Между доходом ушедших и оставшихся клиентов **есть** статистически значимая разница.

Зададим уровень значимости 0.05. 

Имеющиеся характеристики:
- у нас есть две выборки: ушедшие клиенты и оставшиеся
- альтернативная гипотеза двусторонняя
- выборки независимые

Создадим выборки по доходу ушедших клиентов (`data_churn_1`) и оставшихся (`data_churn_0`).

```python
data_churn_0 = data.query('churn==0')['est_salary']
data_churn_1 = data.query('churn==1')['est_salary']
```

Посмотрим на их размер.

```python
data_churn_0.count(), data_churn_1.count()
```

Выборки в несколько раз различаются по количеству клиентов. 

Выберем статистический критерий для проверки гипотезы. Наилучшим образом нам пока подходят два: `t-критерий Стьюдента` и `u-критерий Манна-Уитни`. Посмотрим на них детальнее.

**t-критерий Стьюдента**

Описание:
- метод `scipy.stats.ttest_ind(a, b, equal_var=True/False)`
- применяется для проверки двусторонней гипотезы о равенстве среднего генеральной совокупности некоторому значению
- параметрический

Требования:
- нужна проверка нормальности выборочного среднего
- независимые выборки
- выборки от 30 значений в каждой

**u-критерий Манна-Уитни**

Описание:
- метод `st.mannwhitneyu(a, b)`
- применяется для оценки различий между двумя выборками по признаку, измеренному в количественной или порядковой шкале
- непараметрический (аналог критерия Стьюдента)

Требования:
- не нужно проверки нормальности
- независимые выборки
- выборки от 3 значений в каждой

**Тест Стьюдента**

Прежде чем мы проведем тест, нам нужно проверить нормальность распределения средних в выборке. Для этого используем тест Шапиро-Уилка. Однако нам нужно много выборок, чтобы убедиться в том, что среднее значение в них действительно распределено нормально. Выделим из наших данных 1000 выборок и посчитаем для каждой среднее значение. 

Для этого мы напишем цикл `while`, который запустим 1000 раз. Для этого будем считать с помощью переменной `n` количество итераций цикла. Когда она станет равна 1000 — остановимся.

Внутри цикла мы каждый раз будем случайным образом брать 200 случайных отточных, а затем постоянных клиентов и вычислять их средний доход. Получившийся результат запишем в список.

Зададим уровень значимости alpha=0.05. 

Сформулируем гипотезы: 
- **H0** : если p-value Shapiro < 0.05, выборочные средние в выборках распределены нормально.
- **H1** : если p-value Shapiro > 0.05, выборочные средние в выборках распределены не нормально.

```python
# отточные клиенты

means_1 = []
n = 1
 
while n < 1000:
    n += 1
    sample = data.query('churn==1').sample(200)['est_salary'].mean()
    means_1.append(sample)
    
shapiro(means_1)
```

```python
# постоянные клиенты

means_0 = []
n = 1
 
while n < 1000:
    n += 1
    sample = data.query('churn==0').sample(200)['est_salary'].mean()
    means_0.append(sample)
    
shapiro(means_0)
```

Каждый из тестов раз от раза выдаёт результат p-value больше уровня значимости 0.05, поэтому мы отклоняем нулевую гипотезу и считаем, что выборочные средние в выборках распределены не нормально.

Если бы выборочные средние в выборке были распределены нормально, далее нам нужно было бы узнать, равны ли дисперсии по группам. И в зависимости от этого уточнить дополнительный параметр для `t-теста`. Для этого используем критерий Левена.

Зададим уровень значимости alpha=0.05. 

Сформулируем гипотезы: 
- **H0** : если p-value Levene < 0.05, дисперсии подгрупп однородны.
- **H1** : если p-value Levene > 0.05, дисперсии подгрупп не однородны.

```python
# снова выделим две случайные выборки, которые будем сравнивать между собой

data_churn_0_sample = data.query('churn==0').sample(200)['est_salary']
data_churn_1_sample = data.query('churn==1').sample(200)['est_salary']

levene(data_churn_0_sample, data_churn_1_sample)
```

p-value больше 0.05, поэтому мы принимаем нулевую гипотезу и считаем, что дисперсии подгрупп не однородны. Следовательно при проведении t-test критерий `equal_var` будет со значением `False`.

Исходя из наших данных, мы можем провести оба теста. Но в таком случае нам нужно сделать корректировку уровня значимости, ввиду множественной проверки гипотез. Используем поправку Бонферрони.

```python
# по выборочным данным

alpha = 0.05
alpha_bonf = alpha/2 # поправка Бонферрони

results = st.ttest_ind(data_churn_0, data_churn_1, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha_bonf:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    )
```

**Тест Манна-Уитни**

```python
# по выборочным данным 

alpha = 0.05
alpha_bonf = alpha/2 # поправка Бонферрони

results = st.mannwhitneyu(data_churn_0, data_churn_1)

print('p-значение:', results.pvalue)

if results.pvalue < alpha_bonf:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

**Вывод:** по результатам проверки мы получили разные выводы. \
`t-критерий Стьюдента`: между средним доходом ушедших и оставшихся клиентов нет статистически значимой разницы.\
`u-критерий Манна-Уитни`: между доходом ушедших и оставшихся клиентов есть статистически значимая разница.

Мы доверяем полученому результату `u-критерия Манна-Уитни`, т.к. выборочные средние в выборках не распределены нормально. Проведение теста-Стьюдента было лишь для эксперимента. 

Если же выборочные средние распределены нормально, то стоит использовать `t-критерий Стьюдента`, т.к. у теста бОльшая мощность и меньше вероятность допустить ошибку 2 рода. 

### Гипотеза 2

Сформулируем вторую пару гипотез: 

- **H0** : μ1 = μ2, Между значением кредитного скоринга ушедших и оставшихся клиентов **нет** статистически значимой разницы.
- **H1** : μ1 ≠ μ2, Между значением кредитного скоринга ушедших и оставшихся клиентов **есть** статистически значимая разница. 

Зададим уровень значимости 0.05. 

Имеющиеся характеристики:
- у нас есть две выборки: ушедшие клиенты и оставшиеся
- альтернативная гипотеза двусторонняя
- выборки независимые

Создадим выборки по доходу ушедших клиентов (`data_churn_score_0`) и оставшихся (`data_churn_score_1`).

```python
data_churn_score_0 = data.query('churn==0')['score']
data_churn_score_1 = data.query('churn==1')['score']
```

Посмотрим на их размер.

```python
data_churn_score_0.count(), data_churn_score_1.count()
```

Выборки также в несколько раз различаются по количеству клиентов. 

Как и про проверке 1-й пары гипотез проверим нормальность распределения средних в выборке тестом Шапиро-Уилка, чтобы понять, можем ли мы использовать `t-критерий Стьюдента`.

Зададим уровень значимости alpha=0.05. 

Сформулируем гипотезы: 
- **H0** : если p-value Shapiro < 0.05, средние в выборках распределены нормально.
- **H1** : если p-value Shapiro > 0.05, средние в выборках распределены не нормально.

```python
# отточные клиенты

means_score_1 = []
n = 1
 
while n < 1000:
    n += 1
    sample = data.query('churn==1').sample(200)['score'].mean()
    means_score_1.append(sample)

shapiro(means_score_1)
```

```python
# постоянные клиенты

means_score_0 = []
n = 1
 
while n < 1000:
    n += 1
    sample = data.query('churn==0').sample(200)['score'].mean()
    means_score_0.append(sample)
    
shapiro(means_score_0)
```

При каждом тесте как для отточных, так и для постоянных клиентов получаем p-value больше 0.05, поэтому мы не можем принять нулевую гипотезу и считаем, что средние в выборках распределены не нормально. Поэтому будем использовать `u-критерий Манна-Уитни`.

```python
# по выборочным данным 

alpha = 0.05

results = st.mannwhitneyu(data_churn_score_0, data_churn_score_1)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    )
```

**Вывод:** по результатам проверки `u-критерием Манна-Уитни`, мы выявили, что между кредитным скорингом ушедших и оставшихся клиентов есть статистически значимая разница.

## Шаг 6: Сегментация на основе стратегических показателей

Напишем функцию для вывода сегментов.

```python
# Установка первой строки
first_row_labels = ['Уровень оттока сегмента', 'Количество отточных клиентов', 'Размер сегмента']

def analyze_segment(data, query_condition, first_row_labels):
    
    # Запрос сегмента данных
    segment = (data.query(query_condition).agg({'churn':['mean','sum','count']})).round(2).reset_index()
    
    # Переименование колонок
    segment.rename(columns={'index': 'Критерий', 'churn': 'Показатели'}, inplace=True)
    
    for i, label in enumerate(first_row_labels):
        segment.loc[i, 'Критерий'] = label
        
    return segment
```

### Сегмент 1

- Возраст (`age`): *от 50 до 60 лет.*
- Кредитный скоринг (`score`): *от 820 до 940 баллов.*
- Баллы собственности (`equity`): *более 3 баллов собственности.*

```python
query_condition = '50 <= age <= 60 and equity >= 3 and 820 <= score <= 940'
segment_1 = analyze_segment(data, query_condition, first_row_labels)
segment_1
```

### Сегмент 2

- Возраст (`age`): *от 25 до 35 лет*
- Наличие кредитной карты (`credit_card`): *отсутствует.*
- Последняя активность (`last_activity`): *активный.*

```python
query_condition = '25 <= age <= 35 and credit_card == 0 and last_activity == 1'
segment_2 = analyze_segment(data, query_condition, first_row_labels)
segment_2
```

### Сегмент 3

- Баланс на счёте (`balance`):  *от 750 тыс. и идёт до 3,5 млн. руб.*
- Количество банковских продуктов (`products`): *3 и более продуктов.*

```python
query_condition = 'products >= 3 and 750000 <= balance <= 3000000'
segment_3 = analyze_segment(data, query_condition, first_row_labels)
segment_3
```

### Сегмент 4

- Пол (`gender`): *мужской.*
- Баланс на счёте (`balance`):  *от 750 тыс. и идёт до 3,5 млн. руб.*

```python
query_condition = 'gender == "М" and 750000 <= balance <= 3000000'
segment_4 = analyze_segment(data, query_condition, first_row_labels)
segment_4
```

Преоритизируем сегменты по уровню оттока: 
1. Сегмент 3 — 52%
2. Сегмент 2 — 49%
3. Сегмент 4 — 44%
4. Сегмент 1 — 41%
