### Выполним предобработку данных

#### Приведём названия столбцов в "змеиный-регистр"

Посмотрим на все названия столбцов таблицы data.

```python
data.columns
```

Выбивается только одно значение, приведём его в корректный формат:\
`cityCenters_nearest` > `city_centers_nearest`

```python
data = data.rename(columns={'cityCenters_nearest':'city_centers_nearest'})

# проверим, что теперь всё корректно
data.columns
```

#### Поработаем с пропущенными значениями

Выявим пропущенные значения для каждого столбца и посчитаем их.

```python
data.isna().sum()
```

Сначала поработаем со столбцом `balcony`. Вероятнее всего, если количество балконов в квартире не указано, то их просто нет. Поэтому заполним все пропуски по количеству балконов значением 0 и проверим себя.

```python
data['balcony'] = data['balcony'].fillna(value=0)

data['balcony'].isna().sum()
```

По такому же принципу поработаем со столбцами `parks_around3000` и `ponds_around3000`. 
Вероятнее всего, если число парков и водоёмов в радиусе 3 км не указано, то их просто нет.

Заполним все пропуски по количеству парков и водоёмов значением 0 и проверим себя.

```python
data['parks_around3000'] = data['parks_around3000'].fillna(value=0)

data['parks_around3000'].isna().sum()
```

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(value=0)

data['ponds_around3000'].isna().sum()
```

Теперь поработаем со столбцом `is_apartment`. Выдвинем предположение, если этот пункт не заполнен, то объект недвижимости не имеет отношения к апартаментам (то есть это квартира). В таком случае нам нужно прописать это чётко.\
Заменим пустые клетки значением False и проверим себя.

*Из возможных погрешностей: не все квартировладельцы могут знать, что владеют апартаментами, поэтому и не отмечали данный параметр.*

```python
data['is_apartment'] = data['is_apartment'].fillna(False)

data['is_apartment'].isna().sum()
```

Ещё мы можем поработать со столбцом `locality_name`. Тут все пустые клетки мы можем заменить значением 'Неизвестно', а затем проверим себя.

```python
data['locality_name'] = data['locality_name'].fillna(value='Неизвестно')

data['locality_name'].isna().sum()
```

Поработаем со столбцом `floors_total`, там совсем немного пропусков, удалим их и проверим себя.

```python
data = data.dropna(subset=['floors_total'])

data['floors_total'].isna().sum()
```

И последний столбец с которым мы поработаем - `ceiling_height`. Посмотрим на данные по столбцу.

```python
data['ceiling_height'].describe()
```

У нас есть аномалия со 100-метровым потолком (с ним поработаем позже).

Заменим пустые клетки медианным значением (более достоверно ввиду аномалии выявленной выше) и проверим себя.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())

data['ceiling_height'].isna().sum()
```

Посмотрим в каких столбцах у нас ещё остались пропуски.

```python
data.isna().sum()
```

Пропуски в оставшихся столбцах оставим пустыми, т.к. логически их не заменить и в противном случае эти значения могут исказить выводы. Просто запомним их и учтём в дальнейшем.

В целом пропуски могут иметь как технологическую основу, т.е. возникли во время копирования, считывания или записи данных, изменения формата файла, так и быть вызваны человеческим фактором - кто-то не ввел нужную информацию по незнанию/ невнимательности или просто лени, т.к. критерий на сайте не был обязательным.

#### Поработаем с типом данных

Посмотрим на общую информацию о датафрейме ещё раз.

```python
data.info()
```

Заменим тип данных в столбцах на целочисленный там, где параметр должен быть целым числом. Например, у нас не может быть 3,5 балкона или 2,6 пруда. Напишем цикл.

Также заменим строковый тип данных в столбце `first_day_exposition` на формат даты и заодно придадим лёгкий для восприятия вид.

После округлим все значения до сотых и придадим столбцу с ценой более привычный формат.

```python
cols_to_include = ['balcony','last_price','ponds_around3000','parks_around3000','floors_total','days_exposition']

for col in cols_to_include:
    data[col] = data[data[col].notnull()][col].astype('int')

data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

pd.set_option("display.float_format", "{:.2f}".format)
```

Проверим себя.

```python
data.info()
```

#### Изучим уникальные значения в столбцах и устраним дубликаты

Сначала проверим датасет на явные дубликаты. 

```python
data.duplicated().sum()
```

Всё отлично, идём дальше. Теперь проверим датасет на неявные дубликаты.

Посмотрим на количество уникальных названий населённых пунктов до преобразований.

```python
len(data['locality_name'].unique())
```

Сначала поменяем регистр и сделаем все буквы строчными, чтобы убрать дубликаты такого типа, т.к. они довольно часто встречаются.

```python
data['locality_name'] = data['locality_name'].str.lower()
```

Посмотрим на уникальные названия населённых пунктов.

```python
data['locality_name'].sort_values().unique()
```

Выявляются две проблемы:

1. Есть дубли, различающиеся написанием "ё" и "е" в слове. Заменим все "ё" на "е".


2. Выделяются группы наименований по типам населённых пунктов, а именно:

    - `поселок` - это "поселок", "городской поселок", "поселок городского типа", "поселок при железнодорожной станции", "поселок станции", "коттеджный поселок"
    - `снт`- это "садоводческое некоммерческое товарищество", "садовое товарищество"
    - `деревня`
    - `село`

Столбец "тип этажа квартиры" - `floor_type` со значениями: «первый», «последний», «другой».\
Для этого создадим функцию floor_type()

```python
def floor_type(row):
    floors_total = row['floors_total']
    floor = row['floor']
    try:
        if floor == 1:
            return 'первый'
        elif floor == floors_total:
            return 'последний'
        elif floor != 1 and floor != floors_total:
            return 'другой'
    except:
        return 'другой'
```

Применим функцию и проверим себя.

```python
data['floor_type'] = data.apply(floor_type, axis=1)
data['floor_type'].value_counts()
```

В ходе исследования мы выявили следующее:

- Наибольшее влияние на стоимость квартиры имеет `общий метраж квартиры`, `жилая площадь`, `площадь кухни` и `количество комнат`. Это и логично, с увеличением каждого из параметров растёт и стоимость объекта недвижимости.

- Небольшое влияние оказывает `высота потолков`. Можно предположение, что высота потолков различается в зависимости от времени постройки домов, а они, скорее всего постепенно строились всё дальше и дальше от центра города.
- В свою очередь нами было подтверждено, что чем `ближе  квартира к центру`, тем выше она будет стоить. Ведь там находится исторический центр города и вся основная инфраструктура в пешей доступности.

- Конкретный `этаж`, `день`, `месяц` и `год` размещения объявления о продаже не подтвердили своего существенного влияния, которое мы могли бы перенести на генеральную совокупность. Наши выводы по этим критериям могут остаться только в рамках текущей выборки.
